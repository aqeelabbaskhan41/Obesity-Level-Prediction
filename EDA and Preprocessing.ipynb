## Outlier detection and removal

```python
import scipy.stats as stats

# QQ plot for Age before transformation
plt.figure(figsize=(12, 6))
stats.probplot(df['Age'], dist="norm", plot=plt)
plt.title('QQ Plot of Age (Before Transformation)')
```

Logarithmic transformation on age

```python
df['Age'] = np.log1p(df['Age'])  # Logarithmic transformation
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title('Transformed Age Distribution')
plt.show()
```

```python
# Recalculate outliers for the transformed Age column
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_age_transformed = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]
outlier_count_age_transformed = len(outliers_age_transformed)

print(f"Number of outliers in transformed Age: {outlier_count_age_transformed}")
```

```python
# Remove outliers from the Age column
df = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)].reset_index(drop=True)

# Display the updated DataFrame shape
print(df.shape)
```

```python
# QQ plot for Age after transformation
stats.probplot(df['Age'], dist="norm", plot=plt)
plt.title('QQ Plot of Age (After Transformation)')
plt.tight_layout()
plt.show()

```

```python
Q1 = df['Height'].quantile(0.25)
Q3 = df["Height"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
count = 0
for i in df['Height']:
    if(i<lower_bound or i>upper_bound):
        count= count+1
print(count)
```

```python
Q1 = df['Weight'].quantile(0.25)
Q3 = df["Weight"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
count = 0
for i in df['Weight']:
    if(i<lower_bound or i>upper_bound):
        count= count+1
print(count)
```

```python
Q1 = df['CH2O'].quantile(0.25)
Q3 = df["CH2O"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
count = 0
for i in df['CH2O']:
    if(i<lower_bound or i>upper_bound):
        count= count+1
print(count)
```

## feature construction

```python
# Calculate BMI
df['BMI'] = df['Weight'] / (df['Height'] ** 2)

# Drop the Weight and Height columns
df.drop(columns=['Weight', 'Height'], inplace=True)

df.drop(columns='index', inplace=True)
# Display the updated DataFrame
df.head()
```

```python
# updating the numerical columns list to include BMI
numerical_cols = [
    'Age',
    'BMI',
    'use_vegetables',
    'num_meals_daily',
    'CH2O',
    'physical_activity',
    'use_technological_devices'
]
```

```python
# Use all features from the dataset
X = df.drop(columns=[target_col])  # Exclude the target column
y = df[target_col]  # Target column
```

```python
X.head()
```

Encoding catagorical data

```python
# Perform one-hot encoding on categorical columns
encoded_categorical = pd.get_dummies(X[categorical_cols], drop_first=True, dtype=int)

# Drop the original categorical columns from X
X = X.drop(columns=categorical_cols)
# Concatenate the encoded categorical columns back to X
X = pd.concat([X, encoded_categorical], axis=1)
```

```python
X.head()
```

Numerical Feature Scaling 

```python
# Scale the numerical columns
scaler = StandardScaler()
X[numerical_cols] = scaler.fit_transform(X[numerical_cols])
```

```python
X.head()
```

Label Encoding

```python
# Encode the target column
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```
