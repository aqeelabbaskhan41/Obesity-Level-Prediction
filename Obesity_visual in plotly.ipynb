```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv('/content/drive/MyDrive/ObesityDataSet.csv')
```

```python
df.head()
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df.head()
```

```python
fig = px.scatter(df, x='Age', y='Weight')
fig.show()
```

```python
fig = px.histogram(df, x='food_bw_meals', y='Weight', color='food_bw_meals')
fig.show()
```

```python
fig = px.pie(values=df['food_bw_meals'].value_counts(), names=df['food_bw_meals'].unique())
fig.show()
```

```python
fig = px.histogram(df, x='family_history_with_overweight', color='family_history_with_overweight')
fig.show()
```

```python
fig = px.histogram(df, x='high_caloric_food', color='high_caloric_food')
fig.show()
```

```python
fig = px.pie(values=df['smoking'].value_counts(), names=df['smoking'].unique())
fig.show()
```

```python
fig = px.box(df, x='monitor_num_calories', y='Weight', color='monitor_num_calories')
fig.show()
```

```python
df.head()
```

```python
fig = px.box(df, x='Age', y='alcohol', color='alcohol')
fig.show()
```

```python
fig = px.box(df, x='Age', y='obesity_level', color='obesity_level')
fig.show()
```

```python
fig = px.histogram(df, x='Age', color='obesity_level', barmode='overlay',
                   color_discrete_sequence=px.colors.qualitative.Set1)
# Show the figure
fig.show()
```

```python
sample = df.sample(n=10, random_state=42)
fig = px.line(sample, x='Age', y='Weight')
fig.show()
```

```python
fig = px.strip(df, x='obesity_level', y='Age', color='obesity_level')
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Violin(y=df['Weight'], x=df['traveling'], box_visible=True, meanline_visible=True, points='all', text=df['traveling'], name='Violin'))
fig.update_layout(title='Traveling vs Weight')
fig.show()
```

```python
fig = px.scatter(df, x='Age', y='use_technological_devices', size='Weight', color='obesity_level',
                 hover_name='Gender', log_x=True, size_max=60)
fig.update_layout(title='Age vs Use of Technological Devices with Bubble Size by Weight and Color by Obesity Level')
fig.show()
```

```python
heatmap_data = df.groupby(['obesity_level', 'food_bw_meals']).size().reset_index(name='count')
fig = px.density_heatmap(heatmap_data, x='food_bw_meals', y='obesity_level', z='count')
fig.update_layout(title='Heatmap of Obesity Level vs Food Between Meals')
fig.show()
```
