```python
sample_df=df.sample(n=500, random_state=42)
df1=sample_df[sample_df['Gender']=='Male']
df2=sample_df[sample_df['Gender']=='Female']
```

```python
# Calculate means for each obesity_level
mean_weight = df1.groupby("obesity_level")["Weight"].mean()
mean_height = df1.groupby("obesity_level")["Height"].mean()
mean_age = df1.groupby("obesity_level")["Age"].mean()
mean_activity = df1.groupby("obesity_level")["physical_activity"].mean()

# Create subplot figure
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        "Weight and Obesity",
        "Height and Obesity",
        "Age and Obesity",
        "Physical Activity and Obesity"
    ),
    horizontal_spacing=0.20,
    vertical_spacing=0.30
)

# Barplot 1 - Weight vs Obesity
fig.add_trace(
    go.Bar(x=mean_weight.index, y=mean_weight.values, marker_color='teal', name="Weight"),
    row=1, col=1
)

# Barplot 2 - Height vs Obesity
fig.add_trace(
    go.Bar(x=mean_height.index, y=mean_height.values, marker_color='olive', name="Height"),
    row=1, col=2
)

# Barplot 3 - Age vs Obesity
fig.add_trace(
    go.Bar(x=mean_age.index, y=mean_age.values, marker_color='darkorange', name="Age"),
    row=2, col=1
)

# Barplot 4 - Physical Activity vs Obesity
fig.add_trace(
    go.Bar(x=mean_activity.index, y=mean_activity.values, marker_color='slateblue', name="Physical Activity"),
    row=2, col=2
)

# Update layout
fig.update_layout(
    height=800,
    width=1000,
    title_text='Graph between Float Features and Obesity',
    title_font=dict(size=20, family='Arial Black'),
    showlegend=False
)

# Update axes
for i in range(1, 3):
    for j in range(1, 3):
        fig.update_xaxes(title_text="Obesity Category", tickangle=45, row=i, col=j)

fig.update_yaxes(title_text="Weight", row=1, col=1)
fig.update_yaxes(title_text="Height", row=1, col=2)
fig.update_yaxes(title_text="Age", row=2, col=1)
fig.update_yaxes(title_text="Physical Activity", row=2, col=2)

# Show the figure
fig.show()
```

```python

# Crosstabs
ct1 = pd.crosstab(df1['obesity_level'], df1['family_history_with_overweight'])
ct2 = pd.crosstab(df1['obesity_level'], df1['high_caloric_food'])
ct3 = pd.crosstab(df1['obesity_level'], df1['smoking'])
ct4 = pd.crosstab(df1['obesity_level'], df1['monitor_num_calories'])

# Create subplot figure with 2 rows and 2 columns
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        "Obesity Level vs Family History With Overweight",
        "Obesity Level vs High Caloric Food",
        "Obesity Level vs Smoking",
        "Obesity Level vs Calorie Monitoring"
    ),
    horizontal_spacing=0.15,
    vertical_spacing=0.15
)

# Heatmap 1
fig.add_trace(
    go.Heatmap(
        z=ct1.values,
        x=ct1.columns.tolist(),
        y=ct1.index.tolist(),
        colorscale='Blues',
        showscale=False,
        text=ct1.values,
        hovertemplate="Obesity Level: %{y}<br>Family History: %{x}<br>Count: %{z}<extra></extra>"
    ),
    row=1, col=1
)

# Heatmap 2
fig.add_trace(
    go.Heatmap(
        z=ct2.values,
        x=ct2.columns.tolist(),
        y=ct2.index.tolist(),
        colorscale='Blues',
        showscale=False,
        text=ct2.values,
        hovertemplate="Obesity Level: %{y}<br>High Caloric Food: %{x}<br>Count: %{z}<extra></extra>"
    ),
    row=1, col=2
)

# Heatmap 3
fig.add_trace(
    go.Heatmap(
        z=ct3.values,
        x=ct3.columns.tolist(),
        y=ct3.index.tolist(),
        colorscale='Blues',
        showscale=False,
        text=ct3.values,
        hovertemplate="Obesity Level: %{y}<br>Smoking: %{x}<br>Count: %{z}<extra></extra>"
    ),
    row=2, col=1
)

# Heatmap 4
fig.add_trace(
    go.Heatmap(
        z=ct4.values,
        x=ct4.columns.tolist(),
        y=ct4.index.tolist(),
        colorscale='Blues',
        showscale=False,
        text=ct4.values,
        hovertemplate="Obesity Level: %{y}<br>Calorie Monitoring: %{x}<br>Count: %{z}<extra></extra>"
    ),
    row=2, col=2
)

# Update layout
fig.update_layout(
    height=800,
    width=1000,
    title_text="Obesity Level Crosstab Heatmaps",
    showlegend=False
)

fig.show()
```

```python
# Create subplot grid
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        "Vegetables and Obesity",
        "Meals per Day and Obesity",
        "CH2O and Obesity",
        "Physical Activity and Obesity"
    ),
    horizontal_spacing=0.20,
    vertical_spacing=0.30
)

# Scatterplot 1 - use_vegetables vs obesity_level
fig.add_trace(
    go.Scatter(
        x=df1["obesity_level"],
        y=df1["use_vegetables"],
        mode='markers',
        marker=dict(color='seagreen', size=6, opacity=0.7),
        name="Vegetables"
    ),
    row=1, col=1
)

# Scatterplot 2 - num_meals_daily vs obesity_level
fig.add_trace(
    go.Scatter(
        x=df1["obesity_level"],
        y=df1["num_meals_daily"],
        mode='markers',
        marker=dict(color='darkorange', size=6, opacity=0.7),
        name="Meals per Day"
    ),
    row=1, col=2
)

# Scatterplot 3 - CH2O vs obesity_level
fig.add_trace(
    go.Scatter(
        x=df1["obesity_level"],
        y=df1["CH2O"],
        mode='markers',
        marker=dict(color='dodgerblue', size=6, opacity=0.7),
        name="CH2O"
    ),
    row=2, col=1
)

# Scatterplot 4 - physical_activity vs obesity_level
fig.add_trace(
    go.Scatter(
        x=df1["obesity_level"],
        y=df1["physical_activity"],
        mode='markers',
        marker=dict(color='purple', size=6, opacity=0.7),
        name="Physical Activity"
    ),
    row=2, col=2
)

# Update layout
fig.update_layout(
    height=800,
    width=1000,
    title_text="Scatterplots of Float Features vs Obesity Level",
    title_font=dict(size=18),
    showlegend=False
)

# Update axis labels and ticks
fig.update_xaxes(title_text="Obesity Category", tickangle=45, row=1, col=1)
fig.update_yaxes(title_text="Vegetable Consumption", row=1, col=1)

fig.update_xaxes(title_text="Obesity Category", tickangle=45, row=1, col=2)
fig.update_yaxes(title_text="Number of Meals per Day", row=1, col=2)

fig.update_xaxes(title_text="Obesity Category", tickangle=45, row=2, col=1)
fig.update_yaxes(title_text="CH2O", row=2, col=1)

fig.update_xaxes(title_text="Obesity Category", tickangle=45, row=2, col=2)
fig.update_yaxes(title_text="Physical Activity", row=2, col=2)

# Show the plot
fig.show()
```

```python
# Create 2x2 subplot figure
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=[
        "Weight and Obesity",
        "Height and Obesity",
        "Age and Obesity",
        "Physical Activity and Obesity"
    ]
)

# Helper: mean value per obesity category
weight_avg = df2.groupby("obesity_level")["Weight"].mean()
height_avg = df2.groupby("obesity_level")["Height"].mean()
age_avg = df2.groupby("obesity_level")["Age"].mean()
activity_avg = df2.groupby("obesity_level")["physical_activity"].mean()

# Ensure consistent x-axis order
obesity_categories = weight_avg.index.tolist()

# Barplot 1 - Weight
fig.add_trace(go.Bar(x=obesity_categories, y=weight_avg, name="Weight", marker_color='teal'),
              row=1, col=1)

# Barplot 2 - Height
fig.add_trace(go.Bar(x=obesity_categories, y=height_avg, name="Height", marker_color='olive'),
              row=1, col=2)

# Barplot 3 - Age
fig.add_trace(go.Bar(x=obesity_categories, y=age_avg, name="Age", marker_color='darkcyan'),
              row=2, col=1)

# Barplot 4 - Physical Activity
fig.add_trace(go.Bar(x=obesity_categories, y=activity_avg, name="Physical Activity", marker_color='purple'),
              row=2, col=2)

# Update layout
fig.update_layout(
    height=700, width=1000,
    title_text="Float Features vs Obesity (Female Dataset)",
    showlegend=False,
    title_font=dict(size=20)
)

# Rotate x-axis labels
fig.update_xaxes(tickangle=45)

# Show interactive plot
fig.show()
```
